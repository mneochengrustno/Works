### Граф 1

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

left_nodes = {"V1", "V2", "V3", "V4"}
right_nodes = {"V5", "V6", "V7", "V8"}

edges = [
    ("V1","V2",6),
    ("V1","V5",3),
    ("V1","V6",2),
    ("V5","V6",9),
    ("V6","V2",2),
    ("V6","V7",3),
    ("V2","V7",3),
    ("V2","V3",1),
    ("V7","V3",1),
    ("V7","V8",3),
    ("V8","V4",3),
    ("V4","V3",7),
    ("V8","V3",4),
]

for u, v, w in edges:
    G.add_edge(u, v, weight=w)

pos = {
    "V1": (0, 1),
    "V2": (1, 1),
    "V3": (2, 1),
    "V4": (3, 1),
    "V5": (0, 0),
    "V6": (1, 0),
    "V7": (2, 0),
    "V8": (3, 0)
}

plt.figure(figsize=(7, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="cyan",
    node_size=1500,
    font_size=10,
    width=1.2
)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))

plt.axis("off")
plt.show()
```

```python
#Алгоритм Прима

def Prima(graph, start):
    visited = set([start])
    mst_edges = []
    total_weight = 0

    while len(visited) < len(graph.nodes):
        min_edge = None
        min_weight = float("inf")
        for u in visited:
            for v, attrs in graph[u].items():
                w = attrs['weight']
                if v not in visited and w < min_weight:
                    min_edge = (u, v)
                    min_weight = w

        mst_edges.append((*min_edge, min_weight))
        total_weight += min_weight
        visited.add(min_edge[1])

    return mst_edges, total_weight
```

```python
mst_edges, total_weight = Prima(G, start="V1")

print("Рёбра минимального остова:")
for u, v, w in mst_edges:
    print(f"{u} — {v} : {w}")

print(f"\nОбщий вес остова: {total_weight}")

plt.figure(figsize=(7, 4))

# Все рёбра и вершины
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="cyan",
    edge_color="black",
    node_size=1200,
    font_size=10,
    linewidths=1,
)

mst_graph = nx.Graph()
mst_graph.add_weighted_edges_from(mst_edges)
nx.draw_networkx_edges(
    mst_graph,
    pos,
    edge_color="deeppink",
    width=2.5
)

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="black")

plt.axis("off")
plt.show()
```
